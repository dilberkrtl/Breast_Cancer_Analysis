```python
# Importing necessary libraries and loading the breast cancer dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()  # Loading the breast cancer dataset
```

```python
cancer.keys()
```

```python
# Printing the general description of the dataset
print(cancer['DESCR'])
```

```python
# Importing the train_test_split function to split data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into training and testing sets
# stratify=cancer.target ensures that the data is split in a balanced manner based on the target classes
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)
```

```python
# Lists to store the training and testing accuracies for each number of neighbors
train_accuracy = []  # Training accuracy
test_accuracy = []  # Test accuracy
```

```python
# Defining the range for the number of neighbors (1 to 10)
num_neighbors = range(1, 11)
```

```python
# Importing matplotlib for plotting the graph
import matplotlib.pyplot as plt

# Importing KNeighborsClassifier to create KNN models
from sklearn.neighbors import KNeighborsClassifier

# Looping through different values of n_neighbors (from 1 to 10)
for n_neighbors in num_neighbors:    
    clss = KNeighborsClassifier(n_neighbors)# Creating a KNN model with the current number of neighbors
    clss.fit(x_train, y_train) # Training the model on the training data
    
    # Calculating and storing the accuracy on training and test datasets
    train_accuracy.append(clss.score(x_train, y_train))  # Training accuracy
    test_accuracy.append(clss.score(x_test, y_test))  # Test accuracy

# Plotting the training and test accuracies against the number of neighbors
plt.plot(num_neighbors, train_accuracy, label="Train Accuracy")  # Plotting training accuracy
plt.plot(num_neighbors, test_accuracy, label="Test Accuracy")  # Plotting test accuracy

# Adding labels to the axes and a legend to the graph
plt.xlabel("Neighbors Number")  # Labeling the x-axis as number of neighbors
plt.ylabel("Accuracy")  # Labeling the y-axis as accuracy
plt.legend()  # Displaying the legend
plt.show()  # Showing the plot
```

```python
# Importing KNeighborsRegressor to perform regression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Creating a synthetic dataset for regression using mglearn
x, y = mglearn.datasets.make_wave(n_samples=40)
```

```python
# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
# Creating a KNN regressor with 3 neighbors
reg = KNeighborsRegressor(n_neighbors=3)
```

```python
# Training the regressor on the training data
reg.fit(x_train, y_train)
```

```python
# Calculating the R^2 (R-squared) score on the test data
reg.score(x_test, y_test)
```
